```python
import pymysql
```

```python
conn = pymysql.connect(
    host="localhost",
    user="root",
    password="class1234",
    database="test",
    charset="utf8"
)

cursor = conn.cursor() # 연결된 DB에 SQL을 실행하는 객체

print("MySQL 연결 성공")
```

```python
products = [
    {"name": "키보드", "price": 50000},
    {"name": "마우스", "price": 25000},
    {"name": "모니터", "price": 300000},
    {"name": "노트북", "price": 1200000},
    {"name": "USB", "price": 15000},
    {"name": "웹캠", "price": 80000},
    {"name": "헤드셋", "price": 90000},
    {"name": "스피커", "price": 70000},
    {"name": "외장하드", "price": 140000},
    {"name": "프린터", "price": 220000}
]
```

```python
sql = """
insert into product(name,price)
values(%s, %s)
"""
```

```python
for product in products:
    cursor.execute(sql, (product["name"], product["price"]))
    print ("{i+1}번째 insert 실행")

conn.commit()
```

```python
sql = """  
select * from product
"""

cursor.execute(sql)

rows = cursor.fetchall()
print(type(rows))
print(rows)

for r in rows:
    print(r)

    
```

```python
# id가 1인 모니터의 price null -> 150000 으로 수정
update_price_val = 150000
update_id = 1

sql = """ 
update product
set price = %s
where id = %s
"""

cursor.execute(sql, (150000,1))
conn.commit()
print(f'id {update_id}번 수정완료')
```

```python
try:
    sql = """ 
    delete from product
    where price is null 
    """

    # db null값 연산할때는 db에서 제공하는 null 연산자만을 사용해야한다.
    # is null, is not null,...사용 / =.!= 사용안됨

    cursor.execute(sql)
    conn.commit()
    # select 실행 후 cursor에 반환된 결과셋을 가진다.
    # insert update delete는 cursor에 반환된
    #  행의 수를 가짐
    print(f'{cursor.rowcount}행 삭제 성공')
except:
    print('db 예외 발생')
finally:
    cursor.close()
    conn.close()
    
```
